# HW1_3: 1. Matplotlib
ใช้ Matplotlib ในการสร้างเส้นโค้ง parabola แบบหงายหยาบ ๆ โดยมีรายละเอียดดังนี้

\
1.กำหนดจุด (x, y) จาก np.array สำหรับสร้างเส้น parabola แบบหงายอย่างง่าย\
2.ตั้งชื่อกราฟว่า "2 separated lines"\
3.เส้นซีกซ้ายเป็นสีแดง และเส้นซีกขวาเป็นสีน้ำเงิน

\
image.png

```python
# vertically-oriented parabola(upward) 
# --> y = x^2/4*a

a = 1

x = np.linspace(-10, 10, 100)
y = (x**2) / (4*a)

x_len_half = len(x) // 2
y_len_half = len(y) // 2

plt.figure(figsize=(6, 6))

plt.plot(x[:x_len_half], y[:y_len_half], color='red')
plt.plot(x[x_len_half:], y[y_len_half:], color='blue')

plt.title("2 separated lines")
plt.show()
```

ใช้ Matplotlib ในการสร้างกราฟวงกลมโดยมีรายละเอียดดังนี้

\
1.กำหนดฟังก์ชั่น x, y จาก wave function ที่มีใน numpy library สำหรับการสร้างวงกลม 1 หน่วยย่างง่าย\
2.ตั้งชื่อกราฟว่า "2 separated circle lines"\
3.วงกลมซีกซ้ายเป็นสีเขียว และวงกลมซีกขวาเป็นสีชมพู

\
image.png

```python
# Circle equation for r = 1 and center at (0, 0)
# --> (x - h)^2 + (y - k)^2 = r^2 
# --> x^2 + y^2 = 1
# --> x = cos(theta), y = sin(thetha)

# theta values for the left and right halves
theta_left = np.linspace(np.pi/2, 3*np.pi/2, 100)  # Left half (negative x)
theta_right = np.linspace(-np.pi/2, np.pi/2, 100)  # Right half (positive x)

x_left = np.cos(theta_left)
y_left = np.sin(theta_left)

x_right = np.cos(theta_right)
y_right = np.sin(theta_right)

plt.figure(figsize=(6, 6))

plt.plot(x_left, y_left, color='green')
plt.plot(x_right, y_right, color='pink')

plt.title("2 separated circle lines")
plt.show()
```
