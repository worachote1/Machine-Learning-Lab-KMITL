```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
t = np.linspace(0, 2 * np.pi, 1000)

x = 16 * np.sin(t) ** 3
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)

plt.figure(figsize=(6, 6))

plt.plot(x[x < 0], y[x < 0], color='pink')

plt.plot(x[x > 0], y[x > 0], color='lightskyblue')

plt.title("I love P’Earth")
plt.axis('equal')
plt.show()
```
